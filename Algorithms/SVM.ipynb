# What is a Support Vector Machine?

SVM is a machine learning technique used in the fields of data analysis and pattern recognition. It essentially builds a hyperplane between datasets to calssify data under varying lables. This algorithm can be conceptualized in the form of a graph. Using the dataset, we plot each point on this this graph and have the hyperplane seperate the points into different classes.

Below we will apply this algorithm to a a group of randomly generated points to see how it works.

```python
#our imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
#randomly generate data points
np.random.seed(0)
X = np.r_[np.random.randn(40,2)-[2,2],np.random.randn(40,2)+[2,2]]
Y = [0]*40+[1]*40
```

```python
#fit the model
svm_model = svm.SVC(kernel='linear')
svm_model.fit(X,Y)
```

```python
#get hyperplane
w = svm_model.coef_[0]
a = -w[0]/w[1]
xx = np.linspace(-5,5)
yy = a * xx - (svm_model.intercept_[0]) / w[1]
```

```python
#display output
plt.plot(xx,yy,'k-')

plt.scatter(svm_model.support_vectors_[:,0],
            svm_model.support_vectors_[:,1],s=80,facecolors='none')
plt.scatter(X[:,0],X[:,1],c=Y, cmap=plt.cm.Paired)
plt.axis('tight')
plt.show()
```
